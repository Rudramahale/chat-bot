```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Customer support chat bot\Data\New.csv")
df.head()
```

```python
df.intent.value_counts()
```

```python
df['intent_id'] =df.intent.map({
    "order_status":0,
    "refund_request":1,
    "order_cancellation":2,
    "technical_issue":3
})
df.head()
```

```python
corpus = df["user_message"]
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
v = TfidfVectorizer()
```

```python
tranformed_output = v.fit_transform(corpus)
print(v.vocabulary_)
```

```python
all_feature_name = v.get_feature_names_out()
for word in all_feature_name:
    indx = v.vocabulary_.get(word)
    print(f"{word} {v.idf_[indx]}")
```

```python
x = tranformed_output
y = df['intent_id']
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,y_pred)
```

```python
intent_to_id = {
    "order_status": 0,
    "refund_request": 1,
    "order_cancellation": 2,
    "technical_issue": 3
}

id_to_intent = {v: k for k, v in intent_to_id.items()}
```

```python
import joblib

joblib.dump(model,'../exp_models/intent_classifier.pkl')
joblib.dump(v,'../exp_models/tf-idf_vectorizer2.pkl')
joblib.dump(id_to_intent,'../exp_models/id_to_intent2.pkl')
```
