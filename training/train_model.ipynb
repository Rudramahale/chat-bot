```python
import pandas as pd
```

```python
df = pd.read_csv("D:\Customer support chat bot\Data\intent.csv")
df.head()
```

```python
df.intent.value_counts()
```

```python
df['intent_id'] =df.intent.map({
    "order_status":0,
    "refund_request":1,
    "order_cancellation":2,
    "technical_issue":3
})
df.head()
```

```python
corpus = df["user_message"]
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
v = TfidfVectorizer()
```

```python
tranformed_output = v.fit_transform(corpus)
print(v.vocabulary_)
```

```python
all_feature_name = v.get_feature_names_out()
for word in all_feature_name:
    indx = v.vocabulary_.get(word)
    print(f"{word} {v.idf_[indx]}")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.intent_id, df.user_message,test_size=0.2,random_state=42)
```
